```python
import pandas as pd
data = pd.read_csv(r"H:\学习小分队\机器学习\sklearn-learning\数据预处理和特征工程\Narrativedata.csv"
                  ,index_col=0 #将第一列设置为索引
                  )

data.loc[:,"Age"] = data.loc[:,"Age"].fillna(data.loc[:,"Age"].median())
#.fillna 填空值，在DataFrame里面直接进行填补

data.dropna(axis=0,inplace=True)
#.dropna(axis=0)删除所有有缺失值的行，.dropna(axis=1)删除所有有缺失值的列
#参数inplace，为True表示在原数据集上进行修改，为False表示生成一个复制对象，不修改原数据，默认False

data.info()
```

```python
#将年龄二值化
data_2 = data.copy()
```

```python
#sklearn.preprocessing.Binarizer
#####根据阈值将数据二值化（将特征值设置为0或1）
from sklearn.preprocessing import Binarizer
```

```python
X = data_2.iloc[:,0].values.reshape(-1,1) #类为特征专用，所以不能使用一维数组
transformer = Binarizer(threshold=30).fit_transform(X)
```

```python
transformer
```

```python
#preprocessing.KBinsDiscretizer
######不止分为2类，可以分为多个类别
#这是将连续型变量划分为分类变量的类，能够将连续型变量排序后按顺序分箱后编码。
```

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
X = data.iloc[:,0].values.reshape(-1,1)
est = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
est.fit_transform(X)
```

```python
#查看转换后分的箱：变成了一列中的三箱
set(est.fit_transform(X).ravel())
```

```python
est = KBinsDiscretizer(n_bins=3, encode='onehot', strategy='uniform') #查看转换后分的箱：变成了哑变量
est.fit_transform(X).toarray()
```
